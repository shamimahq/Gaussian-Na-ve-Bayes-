```python
#Separating independant variable and dependent variable("Classes")
X = df.drop(['Classes'], axis=1)
y = df['Classes']
# print(X.head())
print(X.shape)
# print(y.head())
print(y.shape)
```

```python
# Splitting the dataset to Train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 100)

#to know the shape of the train and test dataset.
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# (B) Gaussian Naïve Bayes and performance evaluation without and with parameters optimization

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

GB = GaussianNB()
GB.fit(X_train, y_train)
predictions = GB.predict(X_test)

print(predictions)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions))
```

Here we can see that the accuracy is .86

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))

```

```python
Here we can see that the accuracy is .87
```

# Parameter Optimization

```python
param_grid_nb = {
    'var_smoothing': np.logspace(0,-9, num=100)
}
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
nbModel_grid = GridSearchCV(estimator=GaussianNB(), param_grid=param_grid_nb, verbose=1, cv=4, n_jobs=-1)
nbModel_grid.fit(X_train, y_train)
print(nbModel_grid.best_estimator_)

#Fitting 10 folds for each of 100 candidates, totalling 1000 fits
GaussianNB(priors=None, var_smoothing=1.0)
```

```python
y_pred = nbModel_grid.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred), ": is the confusion matrix")
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred), ": is the accuracy score")
```

Here we can see that the accuracy score has been increased by .90 by optimization.
